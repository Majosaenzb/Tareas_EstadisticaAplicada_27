# Proceso de Medias Móviles MA(2)

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#fijar el tamaño de las graficas
plt.rcParams['figure.figsize'] = [10, 7.5]
```

$Y_t = \varepsilon_t + 0.3 \varepsilon_{t-1}- 0.1 \varepsilon_{t-2}$

```python
np.random.seed(29)
ma2 = np.array([1, 0.3, -0.1])
ar2 = np.array([1])
proceso_MA2 = ArmaProcess(ar2, ma2).generate_sample(1000)
```

```python
#grafica
plt.plot(proceso_MA2)
plt.title(r'Proceso MA(2) $Y_t = \varepsilon_t + 0.3 \varepsilon_{t-1}- 0.1 \varepsilon_{t-2}$')
plt.show()
```

```python
#Prueba raiz unitaria
adf = adfuller(proceso_MA2)
print('p-value: %f' % adf[1])
```

```python
#grafica función de autocorrelación
plot_acf(proceso_MA2, zero=False, lags=30)
plt.title(r'autocorrelación MA(2) $Y_t = \varepsilon_t + 0.3 \varepsilon_{t-1}- 0.1 \varepsilon_{t-2}$ ')
plt.show()
```

```python
#Grafica autocorrelacion parcial
plot_pacf(proceso_MA2, zero=False, lags=30)
plt.title(r'autocorrelación parcial MA(2) $Y_t = \varepsilon_t + 0.3 \varepsilon_{t-1}- 0.1 \varepsilon_{t-2}$')
plt.show()  
```

# Proceso Autorregresivo AR(2)

$Y_t = 0.5 Y_{t-1}+0.4 Y_{t-2}+\varepsilon_t$

```python
np.random.seed(12)
ma2 = np.array([1])
ar2 = np.array([1, -0.5, -0.4])
proceso_AR2 = ArmaProcess(ar2, ma2).generate_sample(1000)
```

```python
#grafica
plt.plot(proceso_AR2)
plt.title(r'Proceso AR(2) $Y_t = 0.5 Y_{t-1}+0.4 Y_{t-2}+\varepsilon_t$')
plt.show()
```

```python
#Prueba raiz unitaria
adf = adfuller(proceso_AR2)
print('p-value: %f' % adf[1])
```

```python
#graficamos función de autocorrelación
plot_acf(proceso_AR2, zero=False, lags=30)
plt.title(r'autocorrelación AR(2) $Y_t = 0.5 Y_{t-1}+0.4 Y_{t-2}+\varepsilon_t$')
plt.show()
```

```python
#Grafica autocorrelacion parcial
plot_pacf(proceso_AR2, zero=False, lags=30)
plt.title(r'autocorrelación parcial AR(2) $Y_t = 0.5 Y_{t-1}+0.4 Y_{t-2}+\varepsilon_t$')
plt.show()  
```
